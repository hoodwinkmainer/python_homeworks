2 задание:

```python
def nod(a, b):
  while b:
    a, b = b, a % b
  return a

def nod4numbers(a, b, c, d):
  nod_ab = nod(a, b)
  nod_abc = nod(nod_ab, c)
  nod_abcd = nod(nod_abc, c)
  return nod_abcd

a = int(input('a = '))
b = int(input('b = '))
c = int(input('c = '))
d = int(input('d = '))
result = nod4numbers(a, b, c, d)
print('НОД этих чисел равен', result)
```

3 задание:

```python
def C(m, n):
    if m == 0 or m == n:
        return 1
    else:
        return C(m, n - 1) + C(m - 1, n - 1)

m = int(input('m = '))
n = int(input('n = '))

result = C(m, n)
print('Биномиальный коэффицент равен', result)
```

```python
class Invoice:
  def __init__(self, number, sum):
    self.number = number
    self.sum = sum

  def displayInfo(self):
    print(f'Номер счёта: {self.number}, Сумма: {self.sum}')

  def update_amount(self, new_amount):
    self.sum = new_amount

account = Invoice (123, 30000)
account.displayInfo()
account.update_amount(40000)
account.displayInfo()
```

```python
class Transaction:
  def __init__(self, num):
    self.num = num
  def displayInfo(self):
    print(f'Номер счёта: {self.num}')

class Payment(Transaction):
  def __init__(self, num, sum):
    super().__init__(num)
    self.sum = sum

  def updateSum(self, newSum):
    self.sum = newSum

  def displayPaymentInfo(self):
    self.displayInfo()
    print(f'Сумма: {self.sum}')

payment = Payment(123123, 10000)
payment.displayPaymentInfo()
payment.updateSum(20000)
payment.displayPaymentInfo()
```

```python
from abc import ABC, abstractmethod
class ChartElement(ABC):
  @abstractmethod
  def draw(self):
    pass

class BarChart(ChartElement):
  def draw(self):
    print('|')

class LineChart(ChartElement):
  def draw(self):
    print('_')

bar = BarChart()
line = LineChart()

bar.draw()
line.draw()
```

# 7 задача

image.png

```python
from abc import ABC, abstractmethod

class IUniversity(ABC):
    @abstractmethod
    def addStudent(self):
        pass

    @abstractmethod
    def getUniversityInfo(self):
        pass

class University(IUniversity):
    def __init__(self, name):
        self.name = name
        self.studentCount = 0

    def addStudent(self):
        self.studentCount += 1
        print(f"Студент добавлен. Теперь количество студентов: {self.studentCount}")

    def getUniversityInfo(self):
        print(f"Университет: {self.name}, Количество студентов: {self.studentCount}")

if __name__ == "__main__":
    uni = University("КФУ")
    uni.addStudent()
    uni.addStudent()
    uni.getUniversityInfo()
```

```python
class Payment:
  def __init__(self, sum):
    self.sum = sum

  def processPayment(self):
    print(f'Sum: {self.sum}')

class creditCardPayment(Payment):
  def processPayment(self):
    print(f'Sum: {self.sum}, Payment method: Credit Card payment')

class payPalPayment(Payment):
  def processPayment(self):
    print(f'Sum: {self.sum}, Payment method: PayPal payment')

payment1 = creditCardPayment(30000)
payment2 = payPalPayment(15000)

payment1.processPayment()
payment2.processPayment()
```

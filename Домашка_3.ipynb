```python
def countMonotonicSegments(arr):
    if not arr:
        return 0
    n = len(arr)
    if n == 1:
        return 1
    count = 1
    trend = 0
    for i in range(1, n):
        if arr[i] > arr[i-1]:
            if trend == -1:
                count += 1
                trend = 1
            elif trend == 0:
                trend = 1
        elif arr[i] < arr[i-1]:
            if trend == 1:
                count += 1
                trend = -1
            elif trend == 0:
                trend = -1
    return count

arr = list(map(int, input('Введите элементы массива через пробел: ').split()))
result = countMonotonicSegments(arr)
print('Количество промежутков монотонности:', result)
```

```python
def removeDuplicates(arr):
    if not arr:
        return []
    result = [arr[0]]
    for i in range(1, len(arr)):
        if arr[i] != arr[i-1]:
            result.append(arr[i])
    return result

arr = list(map(int, input('Введите элементы массива через пробел: ').split()))
result = removeDuplicates(arr)
print('Массив после удаления соседних дубликатов:', result)
```

4 задание:

```python
def calculate_students_stats():
    N = int(input("Введите количество студентов в группе: "))

    failing_students = 0
    total_group_score = 0

    for student in range(1, N + 1):
        print(f"\nВведите 4 оценки студента {student} через пробел:")
        grades = list(map(int, input().split()))

        if len(grades) != 4:
            print("Ошибка! Нужно ввести ровно 4 оценки. Попробуйте снова.")
            return

        average_score = sum(grades) / 4
        total_group_score += average_score

        if average_score < 3:
            failing_students += 1

    if N > 0:
        group_average = total_group_score / N
    else:
        group_average = 0

    print(f"\nКоличество неуспевающих студентов: {failing_students}")
    print(f"Средний балл группы: {group_average:.2f}")

calculate_students_stats()
```

## Dictionary

Задание 1: Подсчет частоты слов: Напишите функцию, которая принимает строку и возвращает словарь, где ключи — это слова, а значения — их частота в строке.

```python
def word_frequency(text):
    frequency = {}
    current_word = []
    for char in text.lower():
        if char.isalnum() or char == "'":
            current_word.append(char)
        else:
            if current_word:
                word = ''.join(current_word)
                frequency[word] = frequency.get(word, 0) + 1
                current_word = []
    if current_word:
        word = ''.join(current_word)
        frequency[word] = frequency.get(word, 0) + 1

    return frequency

text = "Hello world! Hello Python. Python is great, isn't it? World says hello."
print(word_frequency(text))
```

Задание 2: Инвертирование словаря: Напишите функцию, которая принимает словарь и возвращает новый словарь, где ключи и значения поменяны местами. Если значения не уникальны, объедините ключи в список.

```python
def invertDict(originalDict):
    inverted = {}
    for key, value in originalDict.items():
        inverted.setdefault(value, []).append(key)
    for value in inverted:
        if len(inverted[value]) == 1:
            inverted[value] = inverted[value][0]
    return inverted

d = {'a': 1, 'b': 2, 'c': 1, 'd': 3}
print(invertDict(d))
```

Задание 3: Слияние словарей: Напишите функцию, которая принимает два словаря и возвращает новый словарь, содержащий все ключи из обоих словарей. Если ключи совпадают, значения должны быть суммированы.

```python
def mergeDicts(d1, d2):
    result = d1.copy()
    for key in d2:
        result[key] = result.get(key, 0) + d2[key]
    return result

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
print(mergeDicts(dict1, dict2))
```

Задание 4: Фильтрация по значению: Напишите функцию, которая принимает словарь и пороговое значение, и возвращает новый словарь, содержащий только те пары ключ-значение, где значение больше порогового.

```python
def filterDict(d, threshold):
    return {k: v for k, v in d.items() if v > threshold}

data = {'a': 10, 'b': 5, 'c': 15}
print(filterDict(data, 7))
```

Задание 7: Удаление ключей: Напишите функцию, которая принимает словарь и список ключей, и удаляет из словаря все указанные ключи.

```python
def removeKeys(d, keys):
    for key in keys:
        d.pop(key, None)
    return d

user = {'name': 'John', 'age': 30, 'email': 'john@example.com'}
print(removeKeys(user, ['age', 'email']))
```

Задание 8: Группировка по длине ключей: Напишите функцию, которая группирует ключи словаря по их длине и возвращает словарь, где ключи — это длины, а значения — списки ключей соответствующей длины.

```python
def groupKeysByLength(d):
    result = {}
    for key in d:
        length = len(key)
        if length not in result:
            result[length] = []
        result[length].append(key)
    return result

words = {'cat': 1, 'dog': 2, 'elephant': 3, 'bird': 4}
print(groupKeysByLength(words))
```

Задание 9: Поиск ключей по значению: Напишите функцию, которая принимает словарь и значение, и возвращает список всех ключей, связанных с этим значением.

```python
def findKeysByValue(d, value):
    return [k for k, v in d.items() if v == value]

fruit_colors = {'apple': 'red', 'banana': 'yellow', 'cherry': 'red'}
print(findKeysByValue(fruit_colors, 'red'))
```

Задание 10: Объединение значений: Напишите функцию, которая принимает два словаря и объединяет их значения в списки для каждого ключа.

```python
def mergeValues(d1, d2):
    result = {}
    all_keys = set(d1) | set(d2)
    for key in all_keys:
        result[key] = [d1.get(key), d2.get(key)]
    return result

dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
print(mergeValues(dict1, dict2))
```

Задание 1: Создание множества из списка: Напишите функцию, которая принимает список и возвращает множество, содержащее уникальные элементы списка.

```python
def listToSet(lst):
    return set(lst)

print(listToSet([1, 2, 2, 3, 4]))
```

```python
def listToTuple(lst):
    return tuple(lst)

print(listToTuple([1, 2, 3]))
```

Задание 8: Сортировка кортежа: Напишите функцию, которая возвращает отсортированный список на основе элементов кортежа.

```python
def sortTuple(t):
    return tuple(sorted(t))

print(sortTuple((3, 1, 4, 2)))
```

Задание 9: Разбиение кортежа: Напишите функцию, которая разбивает кортеж на две части: первую половину и вторую половину.

```python
def splitTuple(t):
    half = len(t) // 2
    return (t[:half], t[half:])

print(splitTuple((1, 2, 3, 4, 5, 6)))
```

Задание 10: Обратный порядок: Напишите функцию, которая возвращает кортеж с элементами в обратном порядке.

```python
def reverseTuple(t):
    return t[::-1]

print(reverseTuple((1, 2, 3, 4)))
```
